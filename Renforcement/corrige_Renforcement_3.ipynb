#### Initialisation d’un tableau avec des entiers aléatoires
Écrire un programme qui initialise un tableau d’entiers avec n entiers aléatoires compris entre −50 et 50. L’entier n sera demandé à l’utilisateur.

On pensera à importer la bibliothèque adéquate.

```python
from random import *
 
print("Quelle est la taille du tableau ?")
n=int(input())  
t=[]
i=0
while i<n:
    t.append(randint(-50,50))
    i+=1
```

#### Exercice : Construction de tableau des sommes cumulées
Ecrire un programme qui crée un tableau `t`  d'entiers dont la taille est demandée à l'utilisateur,
et dont les cases contiennent la somme cumulée des indices inférieurs ou égaux à celui de la case. La case `t[i]` contiendra la somme `0+1+2+3+...+(i-1)+i`.

```python
print("Quelle est la taille du tableau ?\n");
n=int(input())

t=[]
i=0
somme=0
while i<n :
        somme+=i
        t.append(somme)
        i=i+1
```

#### Exercice : Affichage à partir de la fin 
- Écrire une **fonction**  qui affiche un à un les éléments d'un tableau en partant de la fin.

- L'utiliser pour afficher le tableau `[0,1,5,6,7]` en partant de la fin.

```python
def tabInv(tab):
    i=len(tab)-1
    while (i>=0): 
        print(tab[i])
        i-=1

tabInv([0,1,5,6,7])
```

#### Exercice : Beaucoup de lettres `'a'` dans un tableau 
Écrire une fonction `nba(tab,seuil)` qui prend en paramètre un tableau d de caractères `tab` et un entier positif `seuil` et qui retourne
- `True` si le nombre d'occurences de la lettre `'a'` contenues dans le tableau de caractères `tab` est supérieur ou égal à `seuil`.
- `False` dans le cas contraire

Ecrire une fonction de tests unitaire satisfaits par cette fonction.

```python
def nba(tab, seuil):
    i=0
    resultat=0
    while i<len(tab)  and resultat<=seuil:
        if tab[i]=='a' :
            resultat+=1
        i+=1
    return resultat==seuil
```

#### Exercice : Valeur du dernier entier négatif d'un tableau
Écrire une fonction qui renvoie la valeur du dernier entier **strictement négatif (<0)** contenu dans un tableau d'entiers passé en paramètre ou la valeur 0 si le tableau ne comporte aucune valeur négative.

**Indication** : il pourra être utile de commencer à parcourir le tableau depuis la fin.


```python
def dernierNeg(tab):
    i=len(tab)-1
    while i > -1 and tab[i] >=0:
        i-=1
    if i < 0 :
        return 0
    else :
        return tab[i]
```

#### Exercice : Position d'insertion dans un tableau trié

Ecrire une fonction qui prend en paramètre un flottant `x` et un **tableau trié en ordre croissan**.  La fonction retournera l'indice de la première valeur `x`dans le tableau si elle est présente, ou l'indice de la case  où devrait se trouver `x` dans le tableau après insertion si elle est absente.

Par exemple, si le tableau est `[0, 2.5, 3.6, 5, 6.8, 9, 12]` alors pour `x=9` la fonction retournera `5`,  pour `x=4.2` elle retournera `3` et pour `x=13.7` elle retournera `7`.

```python
def insertionTriee(x, t) :
    i= 0
    while i<len(t) and t[i]<x :
        i=i+1
    return i
```

#### Exercice : Nombre de valeurs impaires contenues dans un tableau
Écrire une fonction qui retourne le nombre de valeurs impaires contenues dans un tableau d'entiers passé en paramètre.

```python
def nbImpaires(tab):
    i=0
    resultat=0
    while i<len(tab):
        if tab[i]%2!=0 :
            resultat=resultat+1
        i=i+1
    return resultat
```

#### Exercice : Nombres premiers contenus dans un tableau
- Ecrire une fonction qui renvoie le nombre d'éléments premiers dans un tableau d'entiers positifs passé en paramètre.

La bibliothèque `sympy` dispose d'une fonction `isprime((nombre)` qui renvoie `Tru` si `nombre` est premier et `False` sinon.

```python
from sympy import *
def nbresPremiers(tab) :
    resultat=0
    i=0
    while i<len(tab) : 
        if isprime(tab[i]):
            resultat=resultat+1
        i=i+1
    return resultat

tab=[1,2,3,5,7,9]
print(nbresPremiers(tab))
```

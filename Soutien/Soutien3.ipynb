# Exercices de révision - 3

#### Initialisation d’un tableau avec des entiers aléatoires
Écrire un programme qui initialise un tableau d’entiers avec n entiers aléatoires compris entre −50 et 50. 

L’entier n sera demandé à l’utilisateur.

On pensera à importer la bibliothèque adéquate.

#### Tableau de multiples de 5
Écrire un programme qui remplit un tableau de n entiers avec des multiples de 5, en commençant par 0,5,10,.... 

La taille du tableau sera demandée à l’utilisateur. 

Le programme affichera le tableau obtenu.

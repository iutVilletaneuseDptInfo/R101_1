#### Question 1 : Saisie d'une planète 
Ecrire une fonction `saisie_planète()` permettant de saisir les informations sur une planète et retournant le dictionnaire obtenu. On veillera à demander le nombre de satellitesjusqu'à ce qu'il soit positif ou nul.

### Exercice 2 : Bibliothèque

On veut créer une base de données correspondant à une biblithèque.

Chaque livre de la bibliothèque est un dictionnaire avec les entrées suivantes:
- un `"titre"` qui est une chaîne de caractères,
- un `"domaine"` qui est une chaîne de caractères,
- un nombre d'`"exemplaires"` dont on contrôlera qu'il est supérieur à 3.


#### Question 1 : Saisie d'un livre 
Ecrire une fonction `saisie_livre()` permettant de saisir les informations sur un livre et retournant le doctionnaire obtenu. On veillera à demander le nombre d'exemplaires jusqu'à ce qu'il soit correct. 

```python
#------------------------------#
#    CORRECTION  ! ! ! ! ! !   #
#------------------------------#

def saisie_livre():
    livre={}
    livre["titre"] = input("titre du livre ? ")
    livre["domaine"] = input("domaine ? ")
    n=-1
    while n<3:
        n = int(input("nb d'exemplaires ? "))
    livre["exemplaires"] = n
    return livre
```

```python
#------------------------------#
#    CORRECTION  ! ! ! ! ! !   #
#------------------------------#

bibliotheque=[]
i=0
while i<5:
    bibliotheque.append(saisie_livre())
    i+=1

print(bibliotheque)
```

### Exercice 3 : Annuaire

On veut créer un annuaire téléphonique.

Chaque personne dans l'annuaire est un dictionnaire avec les entrées suivantes:
- un `"nom"` qui est une chaîne de caractères,
- un numéro `"tel"` de téléphone qui est une chaîne dont on contrôlera avec `tel.isdigit()` qu'elle ne contient que des chiffres.

**NB** `tel.isdigit()` retourne `True` si la chaîne `tel` ne contient que des chiffres et `False` dans le cas contraire.

#### Question 1 : Saisie d'une personne
Ecrire une fonction `saisie_personne()` permettant de saisir les informations sur une personne et retournant le dictionnaire obtenu. On veillera à demander le numéro de téléphone jusqu'à ce qu'il soit correct. 

```python
#------------------------------#
#    CORRECTION  ! ! ! ! ! !   #
#------------------------------#

def saisie_personne():
    personne={}
    personne["nom"] = input("nom de la personne ? ")
    tel=""
    while not tel.isdigit():
        tel = input("numéro de téléphone ? ")
    personne["tel"] = tel
    return personne

print(saisie_personne())
```

#### Question 2 : Saisie de l'annuaire 
En utilisant la fonction `saisie_personne()`, écrire un programme permettant de saisir des informations sur 3 personnes et les stockant dans un tableau `annuaire`.

Afficher l'annuaire contenant l'ensemble des informations.

```python
#------------------------------#
#    CORRECTION  ! ! ! ! ! !   #
#------------------------------#

annuaire=[]
i=0
while i<3:
    annuaire.append(saisie_personne())
    i+=1
print(annuaire)
```

### Exercice 4 : Evolutions du poids des animaux d'un zoo

L'objectif est concevoir un programme permettant
- de saisir et de stocker dans un tableau `zoo` les informations sur l'évolution hebdomadaire du poids de 250  animaux d'un zoo. Les information pour chaque animal seront elle-mêmes stockées dans un dictionnaire. 
- d'afficher ensuite tous les  dictionnaires correspondant aux animaux. 


Les informations sur les animaux seront saisies par l'utilisateur.
Chaque animal aura
- un `"nom"` qui sera une chaîne de caractères,
- un `"tab_poids"` qui sera un tableau de 7 flottants  obtenus lors de 7 pesées quotidiennes (chaque jour de la semaine). 


#### Question 1 : Saisie d'un animal 
Ecrire une fonction `saisie_animal()` permettant la saisie d'un `animal`.

```python
#------------------------------#
#    CORRECTION  ! ! ! ! ! !   #
#------------------------------#

def saisie_animal():
    animal={}
    animal['nom']=input('Donner le nom de l'animal concerné')
    tab_p=[] 
    j=0
    while j<7:
        poids=float(input('poids du jour ?'))
        tab_p.append(poids)
        j+=1
    animal['tab_poids']= tab_p
    return animal 
```

#### Question 2 : Saisie d'un zoo
En utilisant cette fonction, écrire un programme permettant de saisir et stocker dans un  tableau `zoo`   les informations sur 5 animaux.

```python
#------------------------------#
#    CORRECTION  ! ! ! ! ! !   #
#------------------------------#

zoo=[]
i=0
while i< 5 :
    animal = saisie_animal()
    zoo.append(animal)
    i+=1
```

#### Question 3 : Affichage du zoo
Afficher le `zoo` obtenu animal par animal. On affichera le tableau des poids sous le nom de l'animal.

```python
#------------------------------#
#    CORRECTION  ! ! ! ! ! !   #
#------------------------------#

i=0 
while i < len(zoo): 
    print(zoo[i]["nom"])
    print(zoo[i]["tab_poids"])
    i+=1

#Autre méthode    
i=0 
while i < len(zoo): 
    animal=zoo[i]
    print(animal["nom"])
    print(animal["tab_poids"])
    i+=1
```

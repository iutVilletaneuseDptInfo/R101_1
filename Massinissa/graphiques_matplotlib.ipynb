# Graphiques avec matplotlib

Matplotlib permet de tracer des graphiques dans des notebooks.
Ce notebook présente quelques exemples.
Cela nécessite d'importer le module `pyplot` de `matplotlib`.

```python
import matplotlib.pyplot as plt
```

## Premier graphique

Pour tracer une courbe, on utilise `plt.plot` en lui donnant deux tableaux de même taille, l'un contenant les abscisses des points et l'autre les ordonnées.

```python
x = [1,  2, 3, 4,  5, 6]
y = [5, 12, 0, 0, -5, 2]
plt.plot(x, y)
```

## Graphique avec plusieurs courbes

Il est possible de tracer plusieurs courbes sur le même graphique en utilisant plusieurs fois `plt.plot`. On peut aussi préciser une couleur (`color=`) et un label (`label=`) pour faire la légende.

```python

x_rouge = [1, 2, 3, 4, 5, 6]
y_rouge = [1, 4, 0, 2, 1, 4]
plt.plot(x_rouge, y_rouge, color="red", label = "courbe rouge")

x_vert = [1, 2, 3, 4, 3, 2, 1]
y_vert = [2, 4, 4, 2, 0, 0, 2]
plt.plot(x_vert, y_vert, color="green", label="hexagone vert")
plt.title('Exemple de graphique');
plt.legend();
```

## Camembert

Il est possible de tracer un camembert avec `plt.pie`.


```python
x = [1, 2, 3, 4, 10]
plt.pie(x, labels = ['A', 'B', 'C', 'D', 'E'], normalize = True);
```

## Histogrammes

Il est possible de définir des histogrammes avec `plt.hist`.
Il faut donner les valeurs ainsi que les intervalles. Ceux-ci sont définis sous la forme d'un tableau `[v1, v2, v3, ...., vk]` qui donne les intervalles `[v1,v2]`, `[v2,v3]`, ..., `[vk-1, vk]`.

```python
valeurs = [1, 1, 2, 3, 5, 3, 2, 5, 1, 2, 1, 4.8, 5]
intervalles = [0.5, 1.5, 2.5, 3.5, 4.5, 5.5]

plt.hist(valeurs, intervalles, rwidth=0.5);
```

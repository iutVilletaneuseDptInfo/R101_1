# Mastermind

Le mastermind est un jeu où le joueur a 10 essais pour trouver la solution qui consiste en une séquence de 4 couleurs (chaque couleur est une couleur parmi : rouge, jaune, bleu, orange, vert, gris, et la séquence peut contenir plusieurs fois une même couleur).
À chaque essai, le joueur propose une séquence de 4 couleurs et l'ordinateur répond en indiquant le nombre de couleurs bien placées et le nombre de couleurs qui apparaissent dans la solution mais à un autre endroit dans la séquence.

Ce problème consiste à programmer ce jeu. Pour cela, on représentera les couleurs à l'aide de leur initiale ("r" pour rouge, "j" pour jaune, etc) et une séquence de 4 couleurs par une chaîne 4 caractères. Par exemple `"bvvg"` représente la séquence : bleu, vert, vert, gris.

On utilisera la fonction `randint` de la bibliothèque `random` : 

````python
from random import randint
````

## Question 1 

Définir la fonction `isSequence` prenant en paramètre une chaîne de caractères et retournant `True` si celle-ci correspond à une solution possible du jeu, et `False` sinon.
Par exemple, la fonction doit retourner `True` si `"bvvg"` ou `"gjor"` mais `False` pour `"bkvg"` ou `"vvb"`.


````python
##################
#   Correction   #
##################


def isSequence(s):
   if len(s) != 4:
      return False
   couleurs = "rjbovg"
   i = 0
   while i < len(s):
      if s[i] not in couleurs:
         return False
      i += 1   
   return True
````

Vérifiez que votre fonction vérifie les tests unitaires suivants.

````python
def test_isSequence():
    assert isSequence("bvvg")
    assert isSequence("gjor")
    assert not isSequence("bkvg")
    assert not isSequence("vvb")
    print("Test de la fonction isSequence : ok")

test_isSequence()
````

## Question 2

Définir la fonction `randomSolution` prenant retournant une solution aléatoire du Mastermind.

````python
##################
#   Correction   #
##################


def randomSolution():
    couleurs = "rjbovg"
    s = ""
    i = 0
    while i < 4:
        s += couleurs[randint(0, len(couleurs)-1)]
        i += 1
    return s
````

## Question 3

Définir la fonction `inputSequence` demandant à l'utilisateur de saisir une solution possible du Mastermind. La saisie sera répétée jusqu'à ce que l'utilisateur saisisse quelque chose de valide. La saisie sera retournée par la fonction.

````python
##################
#   Correction   #
##################


def inputSequence():
    s = ""
    while not isSequence(s):
        s = input("Saisir une séquence de 4 couleurs")
    return s
````

## Question 4

Définir la fonction `checkSequence` prenant en paramètre une séquence de 4 couleurs et la solution et retournant un tableau de deux cases : 
- la première case contient le nombre de couleurs correctement placées
- la deuxième case contient le nombre de couleurs apparaissant dans la solution mais mal placées.

**Remarque :** cette question n'est pas facile. Pour calculer ces valeurs, il faut compter pour chaque couleur le nombre de fois où elle apparaît dans les deux chaînes et faire la somme. En retranchant le nombre de couleurs bien placées, on obtient le nombre de couleurs mal placées.

````python
##################
#   Correction   #
##################


def checkSequence(sequence, solution):
    couleurs = "rjbovg"
    mal_placees = 0
    i = 0
    while i < len(couleurs):
        mal_placees += min(sequence.count(couleurs[i]), solution.count(couleurs[i]))
        i += 1
    bien_placees = 0
    i = 0
    while i < len(sequence):
        if sequence[i] == solution[i]:
            bien_placees += 1
        i += 1
    return [bien_placees, mal_placees - bien_placees]
````

Vérifiez que votre fonction vérifie les tests unitaires suivants.

````python
def test_checkSequence():
    assert checkSequence("rrrr", "rrrr") == [4, 0]
    assert checkSequence("rbrb", "rrrr") == [2, 0]
    assert checkSequence("rbvg", "gvbr") == [0, 4]
    assert checkSequence("vbvg", "rrbb") == [0, 1]
    assert checkSequence("bggb", "rrbb") == [1, 1]
    assert checkSequence("vgbg", "gvbr") == [1, 2]
    print("Test de la fonction checkSequence : ok")

test_checkSequence()
````

## Question 5

Définir la fonction `printPreviousSequences` qui prend en paramètre un tableau de séquences (celles qui auront déjà été proposées par le joueur) ainsi qu'un tableau où  chaque case contient le nombre de pions bien placés et mal placés, et affiche ces différentes tentatives.

Par exemple, si le premier tableau est : 
```python
["rrrr", "bbbb", "bbgg"]
```
et le deuxième (car la solution est `"bgvb"`) : 
```python
[ [0, 0], [1, 1], [1, 2] ]
```
alors un affichage peut être : 
```
r r r r => BP : 0, MP : 0
b b b b => BP : 1, MP : 1
b b g g => BP : 1, MP : 2
```

````python
##################
#   Correction   #
##################


def strWithSpaces(s):
    if len(s) == 0:
        return ""
    out = s[0]
    i = 1
    while i < len(s):
        out += " " + s[i]
        i += 1
    return out


def printPreviousSequences(sequences, indications):
    i = 0
    while i < len(sequences):
        print(strWithSpaces(sequences[i]), "=> BP :", indications[i][0], "MP :", indications[i][1])
        i += 1
````

````python
printPreviousSequences(["rrrr", "bbbb", "bbgg"], [ [0, 0], [1, 1], [1, 2] ])
````

## Question 6

Définir la fonction `mastermind` permettant de jouer au mastermind.

````python
##################
#   Correction   #
##################


def mastermind():
    sol = randomSolution()
    prev_seq = []
    prev_indications = []

    tentatives = 10

    while tentatives > 0 and (len(prev_seq) == 0 or prev_seq[len(prev_seq)-1] != sol):
        printPreviousSequences(prev_seq, prev_indications)
        print("Proposez une séquence")
        seq = inputSequence()
        prev_seq.append(seq)
        prev_indications.append(checkSequence(seq, sol))
        tentatives -= 1
    if prev_seq[len(prev_seq)-1] == sol:
        print("Gagné !")
    else:
        print("Désolé, vous avez perdu, la séquence à trouver était " + sol)

mastermind()
````

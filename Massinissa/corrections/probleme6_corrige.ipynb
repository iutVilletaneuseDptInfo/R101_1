## Question 1 

Définir la fonction `equipesAleatoires` prenant en paramètre un entier `n` et un nom de fichier d'équipes et retournant un tableau de `n` équipes choisies aléatoirement dans le fichier.
Si le nombre `n` est négatif ou s'il est plus grand que le nombre d'équipes dans le fichier, la fonction retourne un tableau vide.

````python
##################
#   Correction   #
##################


def equipesAleatoires(n, fichier):
    
    if n < 0:
        return []
    
    # ouverture du fichier
    f = open(fichier, "r", encoding="utf-8")
    
    equipes = []
    
    ligne = f.readline() # lire la ligne courante
    while ligne != "":
        equipes.append(ligne.strip())
        ligne = f.readline() # lire la ligne courante
   
    # fermeture du fichier
    f.close()
    
    if n > len(equipes):
        return []
    
    # 0 <= n <= len(equipes)
    res = []
    i = 0
    while i < n:
        # On choisit une équipe aléatoirement
        indice = randint(0, len(equipes)-1)
        res.append(equipes[indice])
        equipes.pop(indice)
        i += 1
    
    return res
    

listeEquipes = equipesAleatoires(4, "equipes.txt")
print(listeEquipes)
````

## Question 2

Définir la fonction `scoresMatch` retournant un tableau contenant les scores des deux équipes disputant le match. Le score est un entier aléatoire compris entre 0 et 4 inclus. La fonction `scoresMatch` prend en paramètre un booléen `ballonOr` qui vaut `True` si c'est un match avec ballon d'or, donc sans ex-aequo possible, et `False` sinon. 

Dans le cas d'un match avec ballon d'or, les scores doivent être tirés aléatoirement jusqu'à ce qu'ils soient différents (pour qu'il n'y ait pas d'ex-aequo).

````python
##################
#   Correction   #
##################


def scoresMatch(ballonOr):
    scores = [randint(0, 4), randint(0, 4)]
    if ballonOr:
        while scores[0] == scores[1]:
            scores = [randint(0, 4), randint(0, 4)]
    return scores

print(scoresMatch(True))
````

## Question 3

Définir la fonction `match` qui prend en paramètre deux équipes (chaînes de caractères), un tableau contenant leur score (comme celui retourné par la fonction précédente), et retourne une chaîne de caractères contenant le nom des équipes et le nombre de buts qu'elles sont marquées au format suivant : 
```python
"eq1 : nbButsEquipe1, eq2 : nbButsEquipe2"
```
où `eq1` et `eq2` sont remplacées par les noms des deux équipes et `nbButsEquipe1` et `nbButsEquipe2` sont des nombres choisis aléatoirement entre 0 et 4 inclus. La valeur retournée par `match("Angleterre", "Pays-Bas")` pourraît être : 
```python
"Angleterre : 2, Pays-Bas : 1"
```

````python
##################
#   Correction   #
##################


def match(eq1, eq2, scores):
    return eq1 + " : " + str(scores[0]) + ", " + eq2 + " : " + str(scores[1])

res = match("Argentine", "Suisse", scoresMatch(False))
print(res)

````

## Question 4

Une poule en football consiste en un match (sans ballon d'or donc avec possibilité d'ex-aequo) entre toute paire d'équipes de la poule.

Définir la fonction `poule` prenant en paramètre un tableau d'équipes et un nom de fichier et écrivrant *à la fin* dans ce fichier le résultats d'une poule entre ces équipes. Si la fonction est appelée avec le tableau `["Italie", "Croatie", "Espagne"]`, alors le fichier généré pourrait être : 
```
POULE
Italie : 3, Croatie : 3
Italie : 2, Espagne : 0
Croatie : 1, Espagne : 1
```

````python
##################
#   Correction   #
##################


def poule(equipes, nomFichier):
    f = open(nomFichier, "a", encoding="utf-8")
    f.write("POULE\n")
    i = 0
    while i < len(equipes):
        j = i + 1
        while j < len(equipes):
            f.write(match(equipes[i], equipes[j], scoresMatch(False)) + "\n")    
            j += 1
        
        i += 1
    
    f.close()
    
    
poule(["France", "Brésil", "Maroc", "Angleterre"], "res_poule.txt")
````

## Question 5

Dans une poule, une équipe gagne 3 points par victoire, 1 point par match nul et 0 point pour une défaite.

Définir la fonction `calculPoints` qui prend en paramètre un tableau de matches, chaque match étant défini par une châine de caractères au format de la question 3 , et retourne un dictionnaire où les clés sont les noms des équipes des différents matches et les valeurs leur nombre de points à l'issue de ces matches.

Par exemple, L'appel de cette fonction avec le tableau
```python
[
    "Italie : 3, Croatie : 3",
    "Italie : 2, Espagne : 0",
    "Croatie : 1, Espagne : 1"
]
```
doit retourner le dictionnaire :
```python
{
    "Italie" : 4,
    "Croatie" : 2,
    "Espagne" : 1
}
```

````python
##################
#   Correction   #
##################


def calculPoints(matches):
    dico = {}
    # pour chaque match
    i = 0
    while i < len(matches):
        m = matches[i]
        tab = m.split(", ")
        
        # On récupère les équipes du match et leur score
        e1 = tab[0].split(" : ")[0]
        s1 = int(tab[0].split(" : ")[1])
        e2 = tab[1].split(" : ")[0]
        s2 = int(tab[1].split(" : ")[1])
        
        # On ajoute les équipes dans le dictionnaire si elles n'y sont pas
        if e1 not in dico:
            dico[e1] = 0
        if e2 not in dico:
            dico[e2] = 0

        # on met à jour leur score
        if s1 > s2:    
            dico[e1] += 3
        elif s1 < s2:
            dico[e2] += 3
        else:
            dico[e1] += 1
            dico[e2] += 1
        
        i += 1
    return dico

print(calculPoints([
    "Italie : 3, Croatie : 3",
    "Italie : 2, Espagne : 0",
    "Croatie : 1, Espagne : 1"
]))
````

## Question 5

Le gagnant d'un d'un ensemble de matches est l'équipe qui a le plus de points. En cas d'égalité, le gagnant est choisi aléatoirement parmi les équipes qui ont le plus de points. 

Définir la fonction `gagnantMatches` prenant en paramètre un dictionnaire d'équipes avec leur points (comme celui retourné par la fonction précédente) et retourne le gagnant.

````python
##################
#   Correction   #
##################


def gagnantMatches(dico):
    
    scoreMax = 0
    equipesGagnantes = []
    
    equipes = list(dico)
    i = 0
    while i < len(dico):
        if dico[equipes[i]] > scoreMax:
            scoreMax = dico[equipes[i]]
            equipesGagnantes = [equipes[i]]
        elif dico[equipes[i]] == scoreMax:
            equipesGagnantes.append(equipes[i])
        i += 1
    return equipesGagnantes[randint(0, len(equipesGagnantes) - 1)]

gagnantMatches({'Italie': 4, 'Croatie': 5, 'Espagne': 5})
````

## Question 6

Définir la fonction `multipoules` prenant en paramètre un tableau d'équipes, un entier `n_pool` et un nom de fichier. La fonction doit créer des poules de taille `n_pool` et générer un fichier contenant le résultat de chaque poule.

**Attention :** La taille du tableau d'équipes doit être un multiple de `n_pool`. Autrement, la fonction affiche un message d'erreur et ne fait rien.

Si la fonction est appelée avec le tableau d'équipes `["Italie", "Croatie", "Espagne", "Brésil", "Suisse", "Portugal"]` et la valeur 3 pour `n_pool`, le fichier générer doit être de la forme : 
```
POULE
Italie : 3, Croatie : 3
Italie : 2, Espagne : 0
Croatie : 1, Espagne : 1
POULE
Brésil : 3, Suisse : 1
Brésil : 4, Portugal : 2
Suisse : 2, Portugal : 3
```


````python
##################
#   Correction   #
##################


def multipoules(equipes, n_pool, nom_fichier):
    
    if len(equipes) % n_pool != 0:
        print("erreur, le nombre d'équipes n'est pas bon")
        return
    
    # supprimer le contenu du fichier s'il existe
    f = open(nom_fichier, "w")
    f.close()
    
    equipes_poule = []
    i = 0
    while i < len(equipes):
        equipes_poule.append(equipes[i])
        if len(equipes_poule) == n_pool:
            poule(equipes_poule, nom_fichier)
            equipes_poule = []
        i += 1
        
equipes = [
    "Italie", "Croatie", "Espagne", 
    "Brésil", "Suisse", "Portugal", 
    "Argentine", "Belgique", "Pays-Bas", 
    "Angleterre", "France", "Maroc"
]
multipoules(equipes, 3, "res_poule.txt")





````

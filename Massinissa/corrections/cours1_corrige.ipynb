## Question 5 : Fonctions de saisie contrôlée

Définir une fonction `saisiePos` retournant un nombre strictement positif. Ce nombre aura été saisi par l'utilisateur et la saisie répétée jusqu'à ce qu'il saisisse un nombre strictement positif.
Modifier alors le programme précédent pour que la valeur de la longueur et de la largeur soient saisies grâce à cette fonction (ne pas hésiter à modifier le message pour la saisie).

```python
##################
#   Correction   #
##################


def saisiePos():
    val = -1
    while val <= 0:
        val = float(input("Valeur strictement positive"))
    return val

print("Longueur :")
longueur = saisiePos()
print("Largeur :")
largeur = saisiePos()
aire = longueur * largeur
if longueur == largeur:
    forme = "carré"
else:
    forme = "rectangle"

print("L'aire de ce", forme, "est de", aire, "m²")
```

Définir une fonction de saisie contrôlée assurant que la valeur saisie est un entier strictement positif **et pair**.

```python
##################
#   Correction   #
##################


def saisiePosPair():
    val = -1
    while val <= 0 or val%2 != 0: #not(val > 0 and val%2 ==0)
        val = int(input("Valeur strictement positive"))
    return val

nb = saisiePosPair()
print("vous avez saisi", nb)
```

Définir une fonction de saisie contrôlée prenant en paramètre un message et un intervalle et retournant un nombre saisi par l'utilisateur qui soit dans cet intervalle.
Le message sera affiché à chaque saisie de l'utilisateur.

```python
##################
#   Correction   #
##################


def saisie(message, a, b):
    if a > b:
        tmp = a
        a = b
        b = tmp
    val = a - 1
    while val < a or val > b:
        val = float(input(message))
    return val
    
long = saisie("Longueur ?", 2, 3)
print("Vous avez saisi", long)
```

## Question 6 : Résultat sous forme d'une chaîne de caractères

Définir la fonction `description` prenant en paramètre une longueur et une largeur et **retournant une chaîne de caractères** décrivant le rectangle. La description est le gabarit suivi de la forme. Par exemple `petit carré`.

Ajouter un docstring pour décrire la fonction.

```python
##################
#   Correction   #
##################


def description(L, l):
    """Retourne la description du rectangle de longueur L et largeur l.
    Cette description est une chaîne de caractères composé du gabarit (petit, 
    moyen, grand) suivi d'un espace puis de la forme (carré ou rectangle).

    :param L: longueur du rectangle en m (Float)
    :param l: largueur du rectangle en m (Float)
    """
    if L == l:
        forme = "carré"
    else:
        forme = "rectangle"
    
    aire = L * l
    if aire < 50:
        gab = "petit"
    elif aire < 100:
        gab = "moyen"
    else:
        gab = "grand"

    return gab + " " + forme


print(description(3.2, 3.2))
print(description(3.2, 32))
print(description(32, 32))
```

Définir une fonction de tests unitaires pour la fonction description.

```python
##################
#   Correction   #
##################


def test_description():
    assert description(3, 20) == "moyen rectangle"
    assert description(1, 5) == "petit rectangle"
    assert description(34, 34) == "grand carré"
    print("Test de la fonction description : ok")

test_description()
```
